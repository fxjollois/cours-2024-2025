```python
import pandas

df = pandas.read_csv("StockEtablissement_utf8_1000.csv", nrows = 1000)
df
```

```python
def f(v, fin = False):
    r = ""
    if not(pandas.isna(v)):
        r += str(v)
        if not(fin):
            r += "+"
    return r
```

```python
adresse = [ f(df["numeroVoieEtablissement"][i]) + f(df["indiceRepetitionEtablissement"][i]) + \
        f(df["typeVoieEtablissement"][i]) + f(df["libelleVoieEtablissement"][i], True) for i in range(df.shape[0])]
```

```python
fichier_a_envoyer = pandas.DataFrame({
    "siren": df["siren"],
    "siret": df["siret"],
    "adresse": adresse,
    "postcode": df["codePostalEtablissement"],
    "city": df["libelleCommuneEtablissement"]
})
fichier_a_envoyer.to_csv("StockEtablissement_utf8_1000__for_search.csv")
```

```python
import requests

url = "https://api-adresse.data.gouv.fr/search/csv/"
fichier = open("StockEtablissement_utf8_1000__for_search.csv", "rb") # ouverture en lecture ("r") binaire ("b")

r = requests.post(url, files = {"data": fichier})
r.text # contenu du résultat dans un format CSV
```

```python
fichier_resultat = open("StockEtablissement_utf8_1000__resultats.csv", "w", encoding = "utf-8")
fichier_resultat.write(r.text)
fichier_resultat.close()
```
